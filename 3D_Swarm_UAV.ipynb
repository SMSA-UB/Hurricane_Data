```python
def new_flock(count, X, Y, Z):
    width = Y - X
    altitude = Z - X
    return (X[:, np.newaxis] + np.random.rand(3, count) * width[:, np.newaxis] 
            + np.random.rand(3, count) * altitude[:, np.newaxis])
```

```python
positions = new_flock(boid_count, np.array([100, 900, 1000]), np.array([200, 1100, 1200]), np.array([200, 1100, 1200]))
positions
```

```python
velocities = new_flock(boid_count, np.array([0, -20, -20]), np.array([10, 20, 20]), np.array([10, 20, 20]))
velocities
```

```python
positions += velocities
```
