```python
import networkx as nx 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

G1=nx.erdos_renyi_graph(10,0.17)
nx.draw(G1, with_labels = True)
```

```python
degree = list()
threshold = 0.01
node_with_iphone = set()
next_iphone = list()

not_iphone_list = set()

for node_temp,temp_degree in G1.degree():
    degree.append(temp_degree)
length = max(degree)

def add_number(num):
    node_with_iphone.add(num)
    run()
    
def no_number(num):
    not_iphone_list.add(num)
```

```python
def run():
    df = pd.DataFrame(0, columns=['Xn' + str(i) for i in range(length+1)], index=[i for i in range(length+1)])
    for ind in range(len(degree)):
        count = 0
        if ind in node_with_iphone or ind in not_iphone_list:
            pass
        else:
            for temp,val in G1.edges(ind):
                if val in node_with_iphone:
                    count +=1
            if degree[ind] != 0:
                if count/degree[ind] >= threshold:
                    next_iphone.append(ind)
            df.iloc[degree[ind],count] += 1
    node_with_iphone.update(next_iphone)
    display(df)
    df.plot(y = ['Xn' + str(i) for i in range(length+1)], kind="bar", figsize=(10,7), width=2)
    plt.xlabel('Degree') 
    plt.ylabel('Number of Nodes with iPhone') 
    plt.title('Social Network Analysis') 
    plt.grid(True) 
    plt.show()
```

```python
add_number(0)
```

```python
X = G1.edges()
x = np.array(X)

a = list()
b = list()
flag = True
for col in x:
    for row in col:
        if row in node_with_iphone:
            if flag:
                a.append(1)
            else:
                b.append(1)
        else:
            if flag:
                a.append(0)
            else:
                b.append(0)
        flag =  not flag

K = pd.DataFrame(np.hstack((x,np.transpose(np.vstack((np.array(a),np.array(b)))))), columns =["starting node"," ending node","iPhone status of starting node","iPhone status of ending node"])
K["empty column 1"] = ""
K["empty column 2"] = ""
K
```
