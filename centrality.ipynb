```python
import pandas as pd
import networkx as nx
input_data = pd.read_csv('network.csv', index_col=0) #enter the csv file with which you want to create a social network
G = nx.DiGraph(input_data.values)
nx.draw(G, with_labels = True)
```

```python
deg_centrality = nx.degree_centrality(G) 
a = deg_centrality
print(a)
max(a.values())
```

```python
closeness_centrality = nx.closeness_centrality(G) 
b = closeness_centrality
print(b)
max(b.values())
```

```python
betweenness_centrality = nx.betweenness_centrality(G, normalized = True) 
c = betweenness_centrality
print(c)
max(c.values())
```

```python
betweenness_centrality = nx.betweenness_centrality(G, normalized = False) 
d = betweenness_centrality
print(d)
max(d.values())
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G) 
e = eigenvector_centrality
print(e)
max(e.values())
```

```python
katz_centrality = nx.katz_centrality(G) 
f = katz_centrality
print(f)
max(f.values())
```

```python
print ("\nThe degree centrality of the network is", a)
print ("\nThe closeness centrality of the network is", b)
print ("\nThe normalized betweenness centrality of the network is", c)
print ("\nThe un-normalized betweenness centrality of the network is", d)
print ("\nThe eigen vector centrality of the network is", e)
print ("\nThe katz centrality of the network is", f)
```
