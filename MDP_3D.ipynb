```python
# Value iteration:

from collections import defaultdict

def make_dict():
	return defaultdict(make_dict)

# Define States:
S = []
for i in range(1,7):
	for j in range(1,7):
		for k in range(1,7):
			if (i,j,k) not in [(2,4,3),(4,2,1)]:
				S.append((i,j,k))

# Define Actions:
A = ['N', 'E', 'W', 'S', 'F', 'B']

# Define state-action pairs (except for terminal states):
P = defaultdict(make_dict)

for s in S:
	if s not in [(6,4,4),(6,6,6)]:
		for a in A:
			P[s][a] = []
			i = s[0]
			j = s[1]
			k = s[2]
			if a == 'N':
				states = [(i,j+1,k),(i+1,j,k),(i-1,j,k)]
			elif a == 'E':
				states = [(i+1,j,k),(i,j+1,k),(i,j-1,k)]
			elif a == 'W':
				states = [(i-1,j,k),(i,j+1,k),(i,j-1,k)]
			elif a == 'S':
				states = [(i,j-1,k),(i+1,j,k),(i-1,j,k)]
			elif a == 'F':
				states = [(i+1,j+1,k+1),(i,j,k),(i,j,k-1)]
			elif a == 'B':
				states = [(i-1,j-1,k+1),(i,j,k),(i,j,k-1)]

			correctStates = []
			for sp in states:
				if sp in S:
					correctStates.append(sp)
				else:
					correctStates.append(s)

			probability = [0.8, 0.1, 0.1]

			for t in range(0,len(correctStates)):
				P[s][a].append((correctStates[t], probability[t], 0)) 	# 0 is the reward


def obtainValues(policy, gamma):
	# initialize V with zeros:
	V = {}
	for s in S:
		V[s] = 0

	while True:
		Vprev = {}
		for s in V:
			Vprev[s] = float(V[s])

		for s in S:

			if s == (6,6,6):
				V[s] = 5
			elif s == (6,4,4):
				V[s] = -5
			else:
				V[s] = 0
				for (a,prob) in policy[s]:
					val = 0
					for (sp, p, r) in P[s][a]:
						val += p*(r + gamma*Vprev[sp])
					V[s] += prob*val

		diff = 0
		for s in V:
			diff += abs(V[s] - Vprev[s])
		if diff <= 10**(-10):
			break

	print(V)
	return V


def obtainPolicy(V, gamma):
	policy = {}

	for s in P:
		policy[s] = []
		actionVal = {}
		maxVal = 0
		for a in P[s]:
			actionVal[a] = 0
			for (sp, p, r) in P[s][a]:
				actionVal[a] += p*(r + gamma*V[sp])

			if actionVal[a] > maxVal:
				maxVal = actionVal[a]

		actions = []
		for a in actionVal:
			if actionVal[a] == maxVal:
				actions.append(a)

		for a in actions:
			policy[s].append((a,1.0/len(actions)))

	return policy


# Define discount:
gamma = 0.7

policy = {}
for s in P:
	policy[s] = [('N',0.35),('E',0.35),('W',0.15),('S',0.15), ('F',0.20), ('B',0.20)]

max_iter = 100000

for i in range(max_iter):
	Values = obtainValues(policy, gamma)
	newPolicy = obtainPolicy(Values, gamma)

	converge = True

	for s in policy:
		if policy[s] != newPolicy[s]:
			converge = False
			break

	if converge == True:
		print ('Policy-Iteration converged at step %d.' %(i+1))
		break

	policy = newPolicy

print(policy)
```
