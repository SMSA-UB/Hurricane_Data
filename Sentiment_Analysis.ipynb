```
import csv

# Import +/- Words
posWords = []
negWords = []
myList = []

with open('positive-words (1).txt', 'r') as csvfile:
	spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
	for row in spamreader:
		if (row[0][0] != '%'):
			posWords.append(str(row[0]))
        
with open('negative-words (1).txt', 'r') as csvfile:
	spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
	for row in spamreader:
		if (row[0][0] != '%'):
			negWords.append(str(row[0]))
   
with open('Tweet_Column.csv') as fo:
  for x in fo.read().split("\n"):
    x = x[1:]
    x = x.replace("," , "")
    x = x.replace('"' , '')
    x = x.strip('\n')
    if (x != '%'):
      myList.append(x)

# print(myList)
i = 1		
pos = 0
neg = 0
net = 0
for x in range(len(myList)):
  myWord = myList[x] 
  myWord = myWord.split(" ")
  tmpPos = 0
  tmpNeg = 0
  # print(myWord)
  if myWord != ['']:
    # print(myWord)
    for Tweet in myWord:
      Tweet = Tweet.lower()
      if (Tweet in posWords):
        tmpPos += 1
          # print("found positive")
      if (Tweet in negWords):
        tmpNeg += 1
          # print("found negative")
    if tmpPos > tmpNeg:
      print("The " + str(i) + " tweet is Positive")
      pos += 1
    elif tmpPos < tmpNeg:
      print("The " + str(i) + " tweet is Negative")
      neg += 1
    else:
      print("The " + str(i) + " tweet is Neutral")
      net += 1
    i = i + 1
  else:
    continue
  
  # print(tmpPos)
  # print(tmpNeg)
print("There are " + str(pos) + " Positive tweets, " + str(neg) + " Negative tweets, " + str(net) + " Neutral tweets.")   
```
